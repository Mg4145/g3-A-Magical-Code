```python
import numpy as np 
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
from statistics import mean
```

```python
#  The convention for naming of csv files is 
# "results_group#_domain#_batch#_nvalue.csv"
```

```python
def get_null_values(group, domain):
    null_avg = []
    for i in range(10, 110, 10): 
        file_path = "results/results_" + group + "_" + domain + "_" + str(i) + ".csv"
        null_avg.append(round(pd.read_csv(file_path)['score'].mean() * 100, 2))
    return null_avg
```

```python
g1_null = get_null_values("1", "nulls")
g2_null = get_null_values("2", "nulls")
g3_null = get_null_values("3", "nulls")
g4_null = get_null_values("4", "nulls")
g5_null = get_null_values("5", "nulls")
g6_null = get_null_values("6", "nulls")
g7_null = get_null_values("7", "nulls")
g8_null = get_null_values("8", "nulls")

n_values = list(range(10, 110, 10))

g1_avg_null = mean(g1_null)
g2_avg_null = mean(g2_null)
g3_avg_null = mean(g3_null)
g4_avg_null = mean(g4_null)
g5_avg_null = mean(g5_null)
g6_avg_null = mean(g6_null)
g7_avg_null = mean(g7_null)
g8_avg_null = mean(g8_null)

null_avg = []
null_avg.append(g1_avg_null)
null_avg.append(g2_avg_null)
null_avg.append(g3_avg_null)
null_avg.append(g4_avg_null)
null_avg.append(g5_avg_null)
null_avg.append(g6_avg_null)
null_avg.append(g7_avg_null)
null_avg.append(g8_avg_null)
```

```python
# Plot values
n_values = list(range(10, 110, 10))
groups = list(range(1, 9))

plt.figure(figsize=(10, 10))
plt.xlabel("n - values")
plt.ylabel("Percentage")
plt.title("All Group's Null Percentage for n-Values")

plt.plot(n_values, g1_null, label = "Group 1")
plt.plot(n_values, g2_null, label = "Group 2")
plt.plot(n_values, g3_null, label = "Group 3")
plt.plot(n_values, g4_null, label = "Group 4")
plt.plot(n_values, g5_null, label = "Group 5")
plt.plot(n_values, g6_null, label = "Group 6")
plt.plot(n_values, g7_null, label = "Group 7")
plt.plot(n_values, g8_null, label = "Group 8")

plt.xticks(n_values)
# plt.yticks()

plt.legend()
plt.show()
```

```python
# Plot values
n_values = list(range(10, 110, 10))
groups = list(range(1, 9))

plt.figure(figsize=(15, 10))
plt.xlabel("Groups")
plt.ylabel("Percentage")
plt.title("All Group's Mean Null Percentage for n-Values")

plt.bar(groups, null_avg, label = "Average")

plt.xticks(groups)
# plt.yticks()

plt.legend()
plt.show()
```

```python
def get_scores(group, batch):
    total_domain_scores = [] # [domain1, domain2, domain3, ... domain8]
    for domain in range(1, 9):
        n_sum = [] # n = 0, 10, 20, ... , 100
        for n in range(0, 110, 10): 
            file_path = "results/results_" + group + "_" + str(domain) + "_" + batch + "_" + str(n) + ".csv"
            try:
                pd.read_csv(file_path)
            except:
                print(f"{file_path}\t File Not Found")
            else: 
                n_sum.append(pd.read_csv(file_path)['score'].sum())
        total_domain_scores.append(round(sum(n_sum), 2))
    return total_domain_scores
```

```python
g1_avg_domain = get_scores("1", "2")
g2_avg_domain = get_scores("2", "2")
g3_avg_domain = get_scores("3", "2")
g4_avg_domain = get_scores("4", "2")
g5_avg_domain = get_scores("5", "2")
g6_avg_domain = get_scores("6", "1")
g7_avg_domain = get_scores("7", "1")
g8_avg_domain = get_scores("8", "2")
```

```python
n_values = list(range(10, 110, 10))
groups = list(range(1, 9))

plt.figure(figsize=(10, 10))
plt.xlabel("Domain")
plt.ylabel("Scores")
plt.title("Scores Over All Shuffles and Domain For Each Group")

plt.plot(groups, g1_avg_domain, label = "Group 1")
plt.plot(groups, g2_avg_domain, label = "Group 2")
plt.plot(groups, g3_avg_domain, label = "Group 3")
plt.plot(groups, g4_avg_domain, label = "Group 4")
plt.plot(groups, g5_avg_domain, label = "Group 5")
plt.plot(groups, g6_avg_domain, label = "Group 6")
plt.plot(groups, g7_avg_domain, label = "Group 7")
plt.plot(groups, g8_avg_domain, label = "Group 8")

plt.scatter(groups, g1_avg_domain)
plt.scatter(groups, g2_avg_domain)
plt.scatter(groups, g3_avg_domain)
plt.scatter(groups, g4_avg_domain)
plt.scatter(groups, g5_avg_domain)
plt.scatter(groups, g6_avg_domain)
plt.scatter(groups, g7_avg_domain)
plt.scatter(groups, g8_avg_domain)

plt.xticks(groups)
# plt.yticks(n_values)

plt.legend()
plt.show()
```

```python
def get_domain_scores(group, domain, batch):
    domain_scores = []
    for n in range(0, 110, 10): 
        file_path = "results/results_" + group + "_" + str(domain) + "_" + batch + "_" + str(n) + ".csv"
        try:
            pd.read_csv(file_path)
        except:
            print(f"{file_path}\t File Not Found")
            domain_scores.append(0)
        else: 
#             df = pd.read_csv(file_path)
#             print(df['score'].sum())
            domain_scores.append(round(pd.read_csv(file_path)['score'].sum(), 2))
            
    return domain_scores
```
